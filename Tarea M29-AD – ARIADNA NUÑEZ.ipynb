```python
#Importacion de sitio web con reques e impresion del contenido (cualquier sitio)
import requests
from bs4 import BeautifulSoup
from pprint import pprint 

ml = 'https://listado.mercadolibre.com.mx/twisted-wonderland_NoIndex_True'
data_ml = requests.get(ml)

#Impresion de la informacion del sitio web
pprint(data_ml.text)
```

```python
#Se separan los headers para una mejor visualizacion del parseo posterior
headers = {
    'Access-Control-Allow-Origin':'*',
    'Access-Control-Allow-Methods':'GET',
    'Access-Contro-Allow-Headers':'Content-Type',
    'Access-Control-Max-Age':'3600',
    'user-Agent': 'Mozilla/5.0(X11); Ubuntu; Linux x36_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

#Se realiza el request sobre el sitio de Scrape this page
sc = 'https://www.scrapethissite.com/pages/simple/'
data_sc = requests.get(sc,headers = headers)

#Variable para realizar parseo y guardar la información
url_sc = []

html = BeautifulSoup(data_sc.text,'html.parser')
countries = html.select('.country')

#Se separan los paises, por nombre, capital, populacion y area
for country in countries:
    name = country.select('.country-name')[0].get_text(strip = True)
    capital = country.select('.country-capital')[0].get_text(strip = True)
    population = country.select('.country-population')[0].get_text(strip = True)
    area = country.select('.country-area')[0].get_text(strip = True)
    url_sc.append({'name':name,'capital':capital,'population':population,'area':area})

pprint(url_sc)
```

```python
#Lista de todos los items dentro de la pgina
page = requests.get(sc,headers)
soup = BeautifulSoup(data_sc.content, 'html.parser')

class_list = set()
tags = {tag.name for tag in soup.find_all()}

#Se iteran todos los tags
for tag in tags:
    #Encontrar todos los elementos del tag:
    for i in soup.find_all():
        if i.has_attr('class'):
            if len(i['class']) !=0:
                class_list.add(''.join(i['class']))
pprint(class_list)
```
